```python
from pdfminer.pdfinterp import PDFResourceManager, process_pdf 
                                                   # 변환해주는걸 실제 메모리에서 작업해주는거         
from pdfminer.converter import TextConverter
                                #PDF 를 Text로 변환
from pdfminer.layout import LAParams
                            #파라미터
from io import StringIO
from io import open
from urllib.request import urlopen

## 순번
# PDFResourceManager  -> 파라미터를 읽어야하므로 LAParams  -> pdf를 텍스트로 변환 -> TextConverter 내용을 실제로 읽어오주는거 -> process_pdf -> 자원해제

```

```python
def readPDF(pdfFile):
    #pdf 리소스 매니저를 생성
    rsrcmgr = PDFResourceManager()
    
    #pdf 내부의 텍스트 입출력을 위한 IO객체를 생성
    retstr = StringIO()
    
    #파라미터 객체
    laparams = LAParams()
    
    #pdf 내용을 텍스트로 변환하기 위한 객체
    device = TextConverter(rsrcmgr, retstr, laparams = laparams)
    
    #pdf 내용을 텍스트로 변환하는 작업
    process_pdf(rsrcmgr, device, pdfFile)
    device.close()
    
    #리턴된 스트링
    content = retstr.getvalue()
    retstr.close()
    
    
    return content

pdfFile = urlopen(\
                 "http://byteofpython-korean.sourceforge.net/byte_of_python.pdf")
outputString = readPDF(pdfFile)
print(outputString)
pdfFile.close()

with open('readPDF.txt','w',encoding="UTF-8") as f:
    ##as f => 자동으로 close() 관리해줌.
    f.write(outputString)
    print("Save PDFile")
```

```python
# <table> 내부의 텍스트를 저장
import csv
from urllib.request import urlopen
from bs4 import BeautifulSoup

html = urlopen(\
              "https://en.wikipedia.org/wiki/Comparison_of_text_editors")

bsObj = BeautifulSoup(html, "html.parser")
# table 태그 중에서 class가 wikitable인 태그들 중 첫번째
table = bsObj.findAll("table",{"class":"wikitable"})[0]

#가져온 table에서 tr만 뽑을 것
rows = table.findAll("tr")
# wt : 텍스트 파일 쓰기 전용
csvFile = open("editors.csv", "wt", newline="",\
              encoding="UTF-8")

#csv 파일 저장 객체

writer = csv.writer(csvFile)
try:
    for row in rows:
        csvRow=[]
        #td, th 태그의 내용을 리스트에 추가
        for cell in row.findAll(["td","th"]):
            csvRow.append(cell.get_text())
        writer.writerow(csvRow)
finally:
    print("csv파일로 저장되었습니다.")
    csvFile.close()
```
