```python
from urllib.request import urlopen
from PIL import Image
from io import BytesIO
from flask import Flask, render_template, request
import os
from werkzeug import secure_filename
import requests
# 로그인 필요한 정보는 못읽어옴
# 보안처리된 것은 오류메시지가 출력됨.
# html = urlopen("http://edu2.kosta.or.kr/assets/images/kosta2.png")
# html = request.urlopen("http://edu2.kosta.or.kr/assets/images/kosta2.png").read()

response = requests.get("http://edu2.kosta.or.kr/assets/images/kosta2.png")
img = Image.open(BytesIO(response.content))

#http responce 객체를 반환받음
print(img)
print(type(img))

print(img.size)

img.save('kosta.jpg')

app=Flask(__name__)

@app.route("/", methods=["GET"])
def upform():
    return render_template('upform.html')

#파일 업로드 처리
@app.route('/fileUpload', methods = ['GET','POST'])
def upload_file():
    # 이미지 업로드 경로를 static 내부로 설정 ***
    UPLOAD_FOLDER = 'static/uploads/'
    app.config['UPLOAD_FOLDER'] = UPLOAD_FOLDER
    if request.method == 'POST' :
        f = request.files['file']
        #저장할 경로 + 파일명
        #file.save(os.path.join(app.config['UPLOAD_Folder'],filename))
        f.save(os.path.join(app.config['UPLOAD_FOLDER'],\
                           secure_filename(f.filename)))
        return render_template('imgview.html',fn=f.filename)

if __name__ == "__main__":
    app.run()
```

```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent

def getUserAgent():
    userAgent = UserAgent()
    header = {'User-Agent':str(userAgent)}
    return header

def getmelon():
    RANK =  54
    req = requests.get("https://www.melon.com/chart/index.htm", \
                       headers = getUserAgent())
    html = req.text
    parse = BeautifulSoup(html,'html.parser')
    titles = parse.find_all("div",{"class":"ellipsis rank01"})
    songs = parse.find_all("div", {"class":"ellipsis rank02"})
    
    title = []
    song= []
    
    for t in titles:
        title.append(t.find('a').text)
        
    for s in songs:
        song.append(s.find('span', {"class" : "checkEllipsis"}).text)
        
    for i in range(RANK):
        print('%3d위: %s - %s'%(i+1, title[i], song[i]))
        
        
if __name__ == "__main__":
    getmelon()
```

```python

import re
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
def getUserAgent():
    userAgent = UserAgent()
    header = {'User-Agent':str(userAgent)}
    return header


top100_url = "https://www.melon.com/chart/index.htm"
html = requests.get(top100_url, headers = getUserAgent())
print(html)
soup = BeautifulSoup(html.text, "html.parser")
#playsong 이 있는 a태그에 접근해서 songId 번호를 가져온다.

for tag in soup.select('#tb_list a[href*=playSong]'):
    title = tag.text
    js = tag['href']
    #크롤링 영상과는 다르게 멜론 사이트의 songId부분에 ''홀따옴표가 사라졌다.
    #따라서 아래와 같이 정규표현식을 수정해주었다.
    matched = re.search(r",(\d+)",js)
    if matched:
        #group(1)으로 괄호친 부부만 가져온다. group(0)는 홑따옴표 전체를 들고 온다.
        song_id = matched.group(1)
        print(matched.group(0),":",matched.group(1))
        song_url = "https://www.melon.com/song/detail.htm?songid="+song_id
        print(title, song_url)
    
    
```
